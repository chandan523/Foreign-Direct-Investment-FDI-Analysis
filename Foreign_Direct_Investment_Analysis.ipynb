```python
# Calculate initial FDI values for each country/region

# Define years
years = range(2000, 2018)

# Generate hypothetical FDI data for India
fdi_data_india = np.random.randint(5000, 20000, size=len(years))

# Generate hypothetical FDI data for United States
fdi_data_us = np.random.randint(10000, 30000, size=len(years))

# Generate hypothetical FDI data for China
fdi_data_china = np.random.randint(8000, 25000, size=len(years))

# Generate hypothetical FDI data for European Union
fdi_data_eu = np.random.randint(12000, 35000, size=len(years))

# Create a DataFrame to store the data
fdi_df = pd.DataFrame({
    'Year': years,
    'India': fdi_data_india,
    'United States': fdi_data_us,
    'China': fdi_data_china,
    'European Union': fdi_data_eu
})

# Set Year column as index
fdi_df.set_index('Year', inplace=True)
```

```python
#Creating Average Exchange Rate list  :- Reference(Rbi website)
Rates = [45.68,47.69,48.39,45.95,44.93,44.27,45.24,40.26,45.99,
         47.44,45.56,47.92,54.40,60.50,61.14,65.46,67.07]
```

**Converting USD '$' to INR '':**

```python
def multiply_columns(df, col_list, num):
    for col in col_list:
        df[col] = df[col] * Rates[col_list.index(col)]/10
    return df
```

```python
FDI_USD= df.copy()
FDI = multiply_columns(df, Year, Rates)
```

```python
#FDI (Amount in  Crores)
FDI.head()
```

**Unpivoting DataFrames from wide to long format:**

```python
# Unpivoting melt Dataframe
melt = pd.melt(FDI_USD, id_vars = Sectors, value_vars = Year, var_name='Year',
    value_name='FDI(US$ Million)',ignore_index=True)
melt
```

```python
#Unpivoting melt_1 Dataframe
melt_1 = pd.melt(FDI, id_vars = Sectors, value_vars = Year, var_name='Year',
    value_name='FDI( Crores)',ignore_index=True)
melt_1=round(melt_1,2)
melt_1
```

```python
# Merging the FDI(USD Million) column of melt Dataframe into melt_1 Dataframe
Merged= melt_1.merge(melt,how='left')
Merged
```

```python
#Sorting the Sectors and Year columns
Sorted = Merged.sort_values(['Sector','Year'], ignore_index=True)
Sorted
```

```python
#Repalcing some Long values of Sector Column to Short form
Sorted = Sorted[['Sector','FDI( Crores)', 'FDI(US$ Million)'
                 ,]].replace(["CONSTRUCTION DEVELOPMENT: Townships, housing, built-up infrastructure and construction-development projects",
                              "SERVICES SECTOR (Fin.,Banking,Insurance,Non Fin/Business,Outsourcing,R&D,Courier,Tech. Testing and Analysis, Other)",
                              'TEA AND COFFEE (PROCESSING & WAREHOUSING COFFEE & RUBBER)'],
                              ["CONSTRUCTION DEVELOPMENT","SERVICES SECTOR",'TEA AND COFFEE'])
```

**Sector-wise Total FDI 2000-17:**

```python
#Grouping by Sector column to find Total FDI Inflow per Sector from FY2000-01 to FY2016-17
Sectorwise_fdi = Sorted.groupby('Sector').sum()
Sectorwise_fdi.sort_values(by='FDI(US$ Million)',ascending=False)
```

**Visualisation:**

```python
Sectorwise_fdi.plot(kind='bar',y='FDI( Crores)',figsize = (25,7), legend= True,
                    title='SECTOR_WISE FDI INFLOWS IN CRORES',
                    ylabel='FDI( Crores)',cmap='summer')
plt.show()
```

**TOP 10 SECTORS Best & Worst Performing Sectors:**


```python
#Top 10 and bottom 10 sectors
Top_10_Sectors = Sectorwise_fdi.nlargest(10,['FDI( Crores)'])
```

```python
#Calculating percentage-wise FDI share among top 10 sectors and among all sectors
Total_fdi = round(melt_1['FDI( Crores)'].sum(),2)
Sum = Top_10_Sectors['FDI( Crores)'].sum()
Top_10_Sectors['In %age'] = round(Top_10_Sectors['FDI( Crores)']/Sum*100,2)
Top_10_Sectors['%age to Total Inflows'] = round((Top_10_Sectors['FDI( Crores)']/Total_fdi)*100,2)
Top_10_Sectors
```

**Visualisation:**

```python
#Creating bar chart to visualise Total FDI inflow in top 10 sectors using Matplotlib

plt.figure(figsize=(15,5))
plt.barh(Top_10_Sectors.index,Top_10_Sectors['FDI( Crores)'],color='Teal')
plt.title('TOP 10 SECTORS')
plt.xlabel('FDI( Crores)')
plt.show()
```

**BOTTOM 10 SECTORS**

```python
#Calculating share among Bottom sectors
Bottom_10_Sectors = Sectorwise_fdi.nsmallest(10,['FDI( Crores)'])
Sum = Bottom_10_Sectors['FDI( Crores)'].sum()
Bottom_10_Sectors['In %age'] = round(Bottom_10_Sectors['FDI( Crores)']/Sum*100,2)
Bottom_10_Sectors['%age to Total Inflows'] = round((Bottom_10_Sectors['FDI( Crores)']/Total_fdi)*100,3)
Bottom_10_Sectors
```

**Visualisation:**

```python
#Creating bar chart to visualise Total FDI inflow in Bottom 10 sectors using Matplotlib

plt.figure(figsize=(10,6))
plt.barh(Bottom_10_Sectors.index,Bottom_10_Sectors['FDI( Crores)'],color = 'Teal')
plt.title('BOTTOM 10 SECTORS')
plt.xlabel('FDI( Crores)')
plt.show()
```
